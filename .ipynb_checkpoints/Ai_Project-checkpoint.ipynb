```python
 def remove_emojis(data) :
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```

```python
 df['tweet']=df['tweet'].apply(remove_emojis)
```

```python
df=df.copy()
def getSubjectivity(text):
     return TextBlob(text).sentiment.subjectivity
    
def getPolarity(text):
    return TextBlob(text).sentiment.polarity


df["sujectivity"]=df["tweet_without_stopwords"].apply(getSubjectivity)
df["polarity"]=df["tweet_without_stopwords"].apply(getPolarity)
```

```python
df.head(50)
```

```python
def getSentimentAnalysis(score):
    if score < 0 :
        return 'Negative'
    elif score == 0:
            return 'Neutral'
    else:
        return 'Positive'
    
df["Analysis"] = df["polarity"].apply(getSentimentAnalysis)
     
```

```python
# Convert Textual data to feature vectors
x=df['tweet_without_stopwords'].values
```

```python
vectorizer = TfidfVectorizer(max_features=100,
                              max_df=0.8,
                              min_df=5,
                             ngram_range=(1,3))

vectors=vectorizer.fit_transform(x)
```

```python
print(vectors)
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
print(feature_names)
```

```python
dense = vectors.todense()
denselist=dense.tolist()
all_keywords=[]
```

```python
for description in denselist:
    i=0
    keywords=[]
    for word in description:
        if word >0:
            keywords.append(feature_names[i])
        i=i+1
    all_keywords.append(keywords)
print (all_keywords[17])
```

```python
from sklearn.model_selection import KFold
kf= KFold(n_splits=20 ,random_state = 22 , shuffle = True )
i = 1
for train_index,test_index in kf.split(df):
    print("Fold",i)
    print ("train data: ", train_index, "\n")
    print (" test data :",test_index)
    i +=1
  
```

```python
Y=df['state']
classifier1 = LogisticRegression()
cross_val_score(classifier1  ,vectors,Y , cv=kf, scoring='accuracy'  , n_jobs= -1 )
```

```python
from sklearn.svm import SVC
Y=df['state']
svm=SVC()
cross_val_score(svm ,vectors,Y , cv=kf, scoring='accuracy'  , n_jobs= -1 )
```

```python
from sklearn.tree import DecisionTreeClassifier
Y=df['state']
classifier3=DecisionTreeClassifier()
cross_val_score(classifier3 ,vectors,Y , cv=kf, scoring='accuracy'  , n_jobs= -1 )
```
